```python
from PIL import Image
from numpy import asarray
import lenna_plugins_py
```

```python
image = Image.open("data/lenna.png")
img = asarray(image)
print(img.shape)
```

```python
image
```

# blur

We can use the blur plugin to blur the image.

```python
config = lenna_plugins_py.blur.default_config()
config['sigma'] = 2.5
```

process the image and display blurred result

```python
blurred = lenna_plugins_py.blur.process(config, img)
blurred_img = Image.fromarray(blurred, 'RGB')
blurred_img
```

# canny
Lets detect edges using canny algorithm.

```python
config = lenna_plugins_py.canny.default_config()
edges = lenna_plugins_py.canny.process(config, img)
edges_img = Image.fromarray(edges, 'RGB')
edges_img
```

# desaturate
Remove some colors.

```python
config = lenna_plugins_py.desaturate.default_config()
desaturated = lenna_plugins_py.desaturate.process(config, img)
desaturated_img = Image.fromarray(desaturated, 'RGB')
desaturated_img
```

# dramatic
Give the image a dramatic touch.
Sources are from [photon-rs](https://crates.io/crates/photon-rs)

```python
config = lenna_plugins_py.dramatic.default_config()
dramatic = lenna_plugins_py.dramatic.process(config, img)
dramatic_img = Image.fromarray(dramatic, 'RGB')
dramatic_img
```

# photon filters
Many filters are available in [photon](https://github.com/silvia-odwyer/photon)

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display

filters = ['cali', 'dramatic', 'firenze', 'golden', 'lix', 'lofi', 'neue', 'obsidian', 'pastel_pink', 'ryo']

def filter_image(filter):
    config = lenna_plugins_py.photon_filters.default_config()
    config['filter'] = filter
    filtered = lenna_plugins_py.photon_filters.process(config, img)
    filtered_img = Image.fromarray(filtered, 'RGB')
    return filtered_img

interact(filter_image, filter=filters)
```
